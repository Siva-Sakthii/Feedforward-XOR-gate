```python
import numpy as np

def darray(n1, n2):
 n3 = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
 for i in range(4):
  for j in range(2):
    if(j==0):
      n3[i][j]=n1[i]
    else:
      n3[i][j]=n2[i]
 return n3


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def NOR_GATE(X_train):
 w0 = 1
 w1 = -2
 w2 = -2

 weighted_sum = w0 + np.dot(X_train[:, 0], w1) + np.dot(X_train[:, 1], w2)

 predictions = sigmoid(weighted_sum)
 for i in range(len(X_train)):
   predictions[i]=np.round(predictions[i])
 return predictions

def AND_GATE(X_train):
 w0 = -3
 w1 = 2
 w2 = 2

 weighted_sum = w0 + np.dot(X_train[:, 0], w1) + np.dot(X_train[:, 1], w2)

 predictions = sigmoid(weighted_sum)
 for i in range(len(X_train)):
   predictions[i]=np.round(predictions[i])
 return predictions

X_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = np.array([[1], [0], [0], [0]])

c=NOR_GATE(X_train)

X_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = np.array([[0], [0], [0], [1]])

d=AND_GATE(X_train)
next = darray(c,d)
predictions = NOR_GATE(next)

print("Input\tPredicted Output")
for i in range(len(X_train)):
    if predictions[i] < 0.5:
        print(X_train[i], "\t", "0")
    else:
      print(X_train[i], "\t", "1")
```
